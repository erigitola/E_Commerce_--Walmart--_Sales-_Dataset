```python
plt.figure(figsize=(10,6))
sns.kdeplot(
    x=df['latitude'],
    y=df['longitude'],
    cmap='Reds',
    fill=True,
    bw_adjust=1
)

plt.title('Density of Store Locations')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

```python
state_counts = df['state'].value_counts()
sns.barplot(x=state_counts.index, y=state_counts.values)
plt.xticks(rotation=45)
plt.title('Number of Stores by State')
plt.xlabel('State')
plt.ylabel('Store Count')
plt.show()
```

```python
import seaborn as sns
```

```python
df['openDate'] = pd.to_datetime(df['openDate'], errors='coerce')
df['openYear'] = df['openDate'].dt.year


sns.histplot(df['openYear'].dropna(), bins=20, kde=False)
plt.title('Store Openings Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Stores Opened')
plt.show()
```

```python
service_cols = [
    'hasGasDepartment', 'hasTiresDepartment', 'hasFoodDepartment',
    'hasHearingDepartment', 'hasPharmacyDepartment', 'hasOpticalDepartment',
    'hasBusinessDepartment', 'hasPhotoCenterDepartment'

]

service_counts = df[service_cols].sum().sort_values(ascending=False)

sns.barplot(x=service_counts.index, y=service_counts.values,palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Service Availabity Across Costco Stores')
plt.ylabel('Number of Stores')
plt.show()
```
